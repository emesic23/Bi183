```python
import numpy as np
import matplotlib.pyplot as plt
```

# Part A:

```python
# Setup variables
N = 100000
n = 1
lam = 40
p = 0.3
```

```python
Z = np.random.negative_binomial(n, p, N)
X1 = Z + np.random.poisson(40, N)
X2 = Z + np.random.poisson(40, N)

corr = np.corrcoef(X1, X2)[0, 1]
print("Correlation for all points: " + str(corr))
```

# Part B:

```python
filter60 = X1 + X2 >= 60

corr60 = np.corrcoef(X1[filter60], X2[filter60])[0, 1]
print("Correlation for X1 + X2 >= 60: " + str(corr60))
plt.scatter(X1, X2)
plt.scatter(X1[filter60], X2[filter60])
```

```python
filter80 = X1 + X2 >= 80

corr80 = np.corrcoef(X1[filter80], X2[filter80])[0, 1]
print("Correlation for X1 + X2 >= 80: " + str(corr80))
plt.scatter(X1, X2, label="Unfiltered (includes all other points)")
plt.scatter(X1[filter60], X2[filter60], label="60 filter (Includes all 80filter points)")
plt.scatter(X1[filter80], X2[filter80], label="80 filter")
plt.xlabel("X1")
plt.ylabel("X2")
plt.legend()
```
